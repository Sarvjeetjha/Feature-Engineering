OHE or Nominal encoding

disadvantage:

1)for feature having many categories , it crete too many new features as the number of category

2)Sparse matrix get created: overfitting

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({'color':['red','green','blue','red','green']})
df
```

```python
#create an instance of one hot encoder
encoder=OneHotEncoder()
#perform fit and transform
encoded=encoder.fit_transform(df)
encoded
```

```python
arr=encoded.toarray()
print(arr)
```

```python
df_encoded=pd.DataFrame(arr,columns=encoder.get_feature_names_out())
df_encoded
```

```python
pd.concat([df,df_encoded],axis=1)
```

```python
encoder.transform([['blue']]).toarray()
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
df_day=pd.DataFrame(df['day'],columns=['day'])
encoded=encoder.fit_transform(df_day)
encoded
```

```python
arr=encoded.toarray()
```

```python
df_encoded=pd.DataFrame(arr,columns=encoder.get_feature_names_out())
df_encoded
```

```python
encoder.transform([['Sun']]).toarray()
```

```python
pd.concat([df[['total_bill','tip','sex','smoker','day']],df_encoded,df[['time','size']]],axis=1)
```
